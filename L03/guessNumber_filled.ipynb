```python
import random

# 1. The computer picks a secret number between 1 and 100
secret_number = random.randint(1, 100)
attempts = 0

print("Welcome to the Number Guessing Game!")
print("I am thinking of a number between 1 and 100. Can you guess it?")

# 2. Start a loop that runs until the player guesses correctly
while True:
    # 3. Get the player's guess
    guess = int(input("Enter your guess: "))
    attempts += 1 

    # 4. Check the guess using conditional statements
    if guess > secret_number:
        print("Too high! Try a lower number.")
    elif guess < secret_number:
        print("Too low! Try a higher number.")
    else:
        print(f"🔢 Awesome job! You guessed {secret_number} in {attempts} attempts!")
        break
    # if attempt >=10, give the player the correct answer
    if attempts >=10:
        print(f"🔢 Out of attempts! The correct number is: {secret_number}")
        break
```
