```python
# Cell 1: Visuals for the game
HANGMAN_STAGES = [
    """
       --------
       |      |
       |      
       |    
       |      
       |     
    - - - - - -
    """,
    """
       --------
       |      |
       |      O
       |    
       |      
       |     
    - - - - - -
    """,
    """
       --------
       |      |
       |      O
       |      |
       |      
       |     
    - - - - - -
    """,
    """
       --------
       |      |
       |      O
       |     /|
       |      
       |     
    - - - - - -
    """,
    """
       --------
       |      |
       |      O
       |     /|\\
       |      
       |     
    - - - - - -
    """,
    """
       --------
       |      |
       |      O
       |     /|\\
       |     / 
       |     
    - - - - - -
    """,
    """
       --------
       |      |
       |      O
       |     /|\\
       |     / \\
       |     
    - - - - - -
    """
]
```

```python
# Cell 2: Game Logic
import random
from IPython.display import clear_output
import sys

def play_hangman():
    # 1. Word pool (feel free to change the words)
    word_pool = ["python", "jupyter", "notebook", "programming", "developer", "algorithm"]
    secret_word = random.choice(word_pool).lower()
    lword_set = set(secret_word)

    # 2. State game variables
    guessed_letters = set()
    wrong_guesses = 0
    max_wrong_guesses = 6
    
    print("Welcome to Hangman! Let's begin.")
    
    # 3. Main Game Loop
    while wrong_guesses < max_wrong_guesses:
        # Clear previous outputs 
        clear_output(wait=True)
        
        # Display the current gallows stage
        print(HANGMAN_STAGES[wrong_guesses])
        
        # Display the hidden word (e.g., p _ t h _ _ )x
        display_word = ""
        for letter in secret_word:
            if letter in guessed_letters:
                display_word += letter
            else: 
                display_word += "_"
        
        # join to add a space between the characters
        print("Word to guess: " + " ".join(display_word))
        print(f"Guessed letters: {', '.join(sorted(guessed_letters)) if guessed_letters else 'None'}")
        print(f"Remaining lives: {max_wrong_guesses - wrong_guesses}\n")
        
        # Force Jupyter to display everything printed above
        sys.stdout.flush()

        # Get player input
        guess = input("Guess a letter: ").lower().strip()

        # create an exit option
        if guess == 'quit':
            clear_output(wait=True)
            print(f"Thanks for playing! 👋 The word was: '{secret_word.upper()}'.")
            break
        
        # Validate input
        if len(guess) != 1 or not guess.isalpha():
            print("Please enter a single valid letter. (Press Enter to continue)")
            input("Press Enter to continue...")
            continue

        # Add guess to the set
        guessed_letters.add(guess)
            
        if guess in guessed_letters:
            print(f"You already guessed '{guess}'. Try a different letter. (Press Enter to continue)")
            input("Press Enter to continue...")
            continue
        
        # Check if guess is wrong
        if guess not in secret_word:
            wrong_guesses += 1
            print(f"Ouch! '{guess}' is not in the word. (Press Enter to continue)")
            input("Press Enter to continue...")
            continue


        if lword_set.issubset(guessed_letters):
            clear_output(wait=True)
            print("🎉 Congratulations! You guessed the word: \'", secret_word.upper(), "\'!")
            break
            
    else:
        # This executes if the while loop finishes without a 'break' (Player lost)
        clear_output(wait=True)
        print(HANGMAN_STAGES[wrong_guesses])
        print(f"💥 Game Over! You ran out of lives. The word was: '{secret_word.upper()}'.")

# Call the function to start the game
play_hangman()
```
