# Introduction to Matplotlib

**Matplotlib** is Python's most popular data visualization library.  
It lets you turn raw numbers into charts, graphs, and plots that are easy to read and share.

With Matplotlib you can:
- **Line charts** — show trends over time
- **Bar charts** — compare categories
- **Scatter plots** — explore relationships between variables
- **Histograms** — see how data is distributed
- **Pie charts** — show proportions
- **Subplots** — display multiple charts in one figure

Matplotlib is the foundation of many other visualization libraries (like **seaborn** and **pandas** plotting).  
Understanding it gives you full control over how your charts look.

## 1. Importing Matplotlib

The main module we use is `matplotlib.pyplot`.  
By convention, it is imported as `plt` — you will see this in virtually every data science project.

We also import **NumPy** (`np`) to help generate sample data.

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

## 2. Your First Plot

The simplest way to make a chart is:
1. Call `plt.plot()` with your data
2. Call `plt.show()` to display it

````python
# x values: 1 through 5
x = [1, 2, 3, 4, 5]
# y values: corresponding heights
y = [2, 4, 1, 5, 3]

plt.plot(x, y)
plt.show()
````

## 3. Anatomy of a Matplotlib Figure

Before going further, it helps to understand the vocabulary Matplotlib uses:

| Term      | Meaning                                               |
|-----------|-------------------------------------------------------|
| **Figure**| The entire image/window that contains everything      |
| **Axes**  | The actual plot area with x-axis, y-axis, and data    |
| **Axis**  | One single axis (x or y) — the number line            |
| **Title** | The label at the top of the plot                      |
| **Label** | The name of an axis (`xlabel`, `ylabel`)              |
| **Legend**| A key that identifies different lines or groups (say we have more than one)      |
| **Tick**  | The marks along an axis (e.g., 0, 10, 20, ...)        |

```
┌──────────── Figure ───────────────────────┐
│  Title                                    │
│  ┌──────── Axes ───────────────────────┐  │
│  │ y-Axis │  *                         │  │
│  │        │    *    *                  │  │
│  │ ylabel │          *                 │  │
│  │        └──────────────────── x-Axis │  │
│  │                xlabel               │  │
│  └─────────────────────────────────────┘  │
└───────────────────────────────────────────┘
```

## 4. Adding Labels and Titles

A chart without labels is hard to interpret. Always add:
- A **title** — what the chart is about
- **x-axis label** — what the horizontal axis represents
- **y-axis label** — what the vertical axis represents

note: Matplotlib automatically creates both the figure and the axes for you when you call plt.plot().

````python
x = [1, 2, 3, 4, 5]
y = [10, 24, 18, 35, 30]

plt.plot(x, y)

plt.title("Monthly Sales")       # title at the top
plt.xlabel("Month")              # label on the x-axis
plt.ylabel("Revenue ($)")        # label on the y-axis

plt.show()
````

## 5. Line Charts

A **line chart** connects data points with a line.  
Best used when data changes over time (trends, sequences).

You can control the **color**, **line style**, and **marker** shape.  

syntax:
```
plt.plot(
    x, y,       # other than x and y, others are all optional
    color=None,
    linewidth=None,
    linestyle=None,
    marker=None,
    markersize=None,
    label=None,
    alpha=None,
    markerfacecolor=None,
    markeredgecolor=None,
    markeredgewidth=None,
    zorder=None
)
```

| Parameter           | Description        | Example         |
| ------------------- | ------------------ | --------------- |
| `x`                 | x-axis data        | `[1,2,3,4]`     |
| `y`                 | y-axis data        | `[10,20,30,40]` |
| `color` (`c`)       | Line color         | `"red"`         |
| `linewidth` (`lw`)  | Line thickness     | `2`             |
| `linestyle` (`ls`)  | Line style         | `"--"`          |
| `marker`            | Marker style       | `"o"`           |
| `markersize` (`ms`) | Marker size        | `8`             |
| `label`             | Legend label       | `"Sales"`       |
| `alpha`             | Transparency (0–1) | `0.7`           |
| `zorder`            | Layer order        | `3`             |


````python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales  = [200, 240, 180, 310, 290, 350]

plt.plot(months, sales, color="steelblue", linewidth=2, marker="o", linestyle="--")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
````

````python
# Multiple lines on the same chart
math = [72, 80, 68, 90, 85, 95]
reading = [60, 75, 80, 70, 88, 82]

plt.plot(months, math,    label="Math",    marker="o")
plt.plot(months, reading, label="Reading", marker="s", linestyle="--")

plt.title("Scores Over Time")
plt.xlabel("Month")
plt.ylabel("Score")
plt.legend() # shows the label for each line
plt.show()
````

**Common marker styles:**

| Marker | Shape        |
|--------|--------------|
| `"o"`  | Circle       |
| `"s"`  | Square       |
| `"^"`  | Triangle up  |
| `"D"`  | Diamond      |
| `"*"`  | Star         |
| `"+"`  | Plus         |

**Common line styles:**

| Style       | Appearance        |
|-------------|-------------------|
| `"-"`       | Solid (default)   |
| `"--"`      | Dashed            |
| `":"`       | Dotted            |
| `"-."`      | Dash-dot          |

Try different color, line style, markers on your own!

## 6. Bar Charts

A **bar chart** uses rectangular bars to compare values across categories.  
Best used when comparing a measurement across different groups.

````python
subjects = ["Math", "Reading", "Writing", "Science"]
avg_scores = [72, 80, 75, 68]

plt.bar(subjects, avg_scores, color="steelblue")

plt.title("Average Score by Subject")
plt.xlabel("Subject")
plt.ylabel("Average Score")
plt.show()
````

````python
# Horizontal bar chart
plt.barh(subjects, avg_scores, color="coral")

plt.title("Average Score by Subject")
plt.xlabel("Average Score")
plt.show()
````

## 7. Scatter Plots

A **scatter plot** shows individual data points on an x-y grid.  
Best used to explore the **relationship** (correlation) between two numeric variables.

- Points sloping upward - positive correlation
- Points sloping downward - negative correlation
- No pattern - no correlation

````python
# Generate sample data
np.random.seed(42)
math_scores = np.random.randint(40, 100, 100)
reading_scores = math_scores + np.random.randint(-15, 15, 100)  # correlated

plt.scatter(math_scores, reading_scores, alpha=0.6, color="steelblue")

plt.title("Math vs Reading Scores")
plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.show()
````

````python
import random 

# Freshmen students
freshman_hours = x = list(range(1, 21)) # 1 to 20
freshman_scores = [random.randint(0, 100) for _ in range(20)]

# Senior students
senior_hours = list(range(1, 21)) # 1 to 20
senior_scores = [random.randint(0, 100) for _ in range(20)]

plt.scatter(freshman_hours,freshman_scores,color="steelblue",label="Freshman",alpha=0.7)

plt.scatter(senior_hours,senior_scores,color="coral",label="Senior",alpha=0.7)

plt.title("Study Hours vs Exam Score")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.legend()

plt.show()
````

## 8. Histograms

A **histogram** shows the **distribution** of a single numeric variable.  
It groups values into bins and shows how many data points fall in each bin.

Unlike a bar chart (which compares categories), a histogram shows the **shape** of your data:
- Is it symmetric or skewed?
- Is there one peak or multiple?
- Are there outliers?

````python
scores = np.random.normal(loc=70, scale=10, size=200)  # mean=70, std=10

plt.hist(scores, bins=20, color="steelblue", edgecolor="white")

# bins= controls how many groups (bars) to divide the data into
# edgecolor adds a white border between bars for clarity

plt.title("Distribution of Test Scores")
plt.xlabel("Score")
plt.ylabel("Number of Students")
plt.show()
````

````python
# Overlapping histograms to compare two groups
senior_hours_scores   = np.random.normal(70, 10, 100)
freshman_scores = np.random.normal(75, 8,  100)

plt.hist(senior_hours_scores,   bins=20, alpha=0.6, label="senior",   color="steelblue")
plt.hist(freshman_scores, bins=20, alpha=0.6, label="freshman", color="coral")

plt.title("Score Distribution by grade")
plt.xlabel("Score")
plt.ylabel("Count")
plt.legend()
plt.show()
````

## 9. Pie Charts

A **pie chart** shows how a whole is divided into parts (proportions).  
Best used when you have a small number of categories that add up to 100%.

note: Pie charts can be hard to read when there are many slices. Use a bar chart instead if you have more than 5–6 categories.

````python
groups  = ["Group A", "Group B", "Group C", "Group D", "Group E"]
counts  = [89, 90, 196, 262, 190]

plt.pie(
    counts,
    labels=groups,
    autopct="%1.1f%%",   # show percentage on each slice
)

plt.title("Students by Ethnicity Group")
plt.show()
````

````python
# explode= how far each slice should be moved away from the center of the pie chart.
explode = [0, 0, 0.1, 0, 0]   # pull out Group C

plt.pie(
    counts,
    labels=groups,
    autopct="%1.1f%%",
    startangle=90,
    explode=explode
)

plt.title("Students by Ethnicity Group (Group C Highlighted)")
plt.show()
````

## 10. Box Plots

A **box plot** (also called a box-and-whisker plot) summarizes the distribution of data using 5 statistics:

| Part           | Meaning                          |
|----------------|----------------------------------|
| Bottom whisker | Minimum (excluding outliers)     |
| Bottom of box  | 25th percentile (Q1)             |
| Line in box    | Median (50th percentile)         |
| Top of box     | 75th percentile (Q3)             |
| Top whisker    | Maximum (excluding outliers)     |
| Dots (○)       | Outliers                         |

Box plots are great for **comparing distributions** across groups.

````python
# Single box plot
scores = np.random.normal(70, 12, 200)

plt.boxplot(scores)
plt.title("Math Score Distribution")
plt.ylabel("Score")
plt.xticks([1], ["Math"])   # label the box
plt.show()
````

````python
# Multiple box plots side by side
math    = np.random.normal(70, 10, 200)
reading = np.random.normal(75, 8,  200)
writing = np.random.normal(72, 11, 200)

plt.boxplot([math, reading, writing], labels=["Math", "Reading", "Writing"])

plt.title("Score Distributions by Subject")
plt.ylabel("Score")
plt.show()
````

## 11. Customizing Your Plot

Matplotlib gives you precise control over how every element looks.  
Here are the most commonly used customizations.

````python
x = [1, 2, 3, 4, 5]
y = [10, 24, 18, 35, 30]

plt.plot(x, y, color="darkorange", linewidth=2.5, linestyle="--", marker="D", markersize=8)

# Set axis limits
plt.xlim(0, 6)        # x-axis from 0 to 6
plt.ylim(0, 40)       # y-axis from 0 to 40

# Add a grid for readability
plt.grid(True, linestyle=":", alpha=0.7)

# Customize tick marks
plt.xticks([1, 2, 3, 4, 5], ["Mon", "Tue", "Wed", "Thu", "Fri"])

# Add a text annotation directly on the plot
# xy = where the arrow points
# xytext = where the text shows
plt.annotate("Peak!", xy=(4, 35), xytext=(3.2, 36),
             arrowprops=dict(arrowstyle="->", color="black"))

plt.title("Customized Line Chart", fontsize=14, fontweight="bold")
plt.xlabel("Day of Week")
plt.ylabel("Value")
plt.show()
````

You can explore more on this [here]("https://matplotlib.org/stable/users/explain/customizing.html")

## 12. Figure Size and DPI

By default, plots are small. You can control the size using `plt.figure()` before plotting.  
`figsize=(width, height)` is in **inches**. `dpi` controls resolution (how many pixels are drawn per inch of your figure).

So Matplotlib converts: figure size (in inches) × dpi = final image resolution (in pixels)

````python
# Create a wider figure before plotting
plt.figure(figsize=(10, 4), dpi=100)   # 10 inches wide, 4 inches tall, 100 pixels per inch

x = np.linspace(0, 10, 100)   # 100 evenly spaced points from 0 to 10
y = np.sin(x)

plt.plot(x, y, color="steelblue")
plt.title("Sine Wave (Wide Figure)")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
````

## 13. Subplots

**Subplots** let you display multiple charts in a single figure, arranged in a grid.  
`plt.subplots(rows, cols)` returns a Figure and an array of Axes (individual plot areas).

````python
# 1 row, 2 columns - two charts side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

x = [1, 2, 3, 4, 5]
y = [10, 24, 18, 35, 30]

# Left plot
axes[0].plot(x, y, color="steelblue", marker="o")
axes[0].set_title("Line Chart")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")

# Right plot
axes[1].bar(x, y, color="coral")
axes[1].set_title("Bar Chart")
axes[1].set_xlabel("x")
axes[1].set_ylabel("y")

plt.tight_layout()   # prevent plots from overlapping
plt.show()
````

````python
# 2 rows, 2 columns - a 2x2 grid of charts
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

scores = np.random.normal(70, 10, 200)

# Top-left: line chart
axes[0, 0].plot(np.sort(scores), color="steelblue")
axes[0, 0].set_title("Sorted Scores (Line)")

# Top-right: histogram
axes[0, 1].hist(scores, bins=20, color="mediumseagreen", edgecolor="white")
axes[0, 1].set_title("Score Distribution (Histogram)")

# Bottom-left: bar chart
groups = ["A", "B", "C", "D"]
vals   = [72, 80, 68, 75]
axes[1, 0].bar(groups, vals, color="coral")
axes[1, 0].set_title("Avg Score by Group (Bar)")

# Bottom-right: scatter plot
y2 = scores + np.random.randint(-10, 10, 200)
axes[1, 1].scatter(scores, y2, alpha=0.5, color="slateblue")
axes[1, 1].set_title("Score Correlation (Scatter)")

plt.suptitle("Student Performance Dashboard", fontsize=16, fontweight="bold")
plt.tight_layout()
plt.show()
````

## 14. Styles and Themes

## 15. Saving a Figure

Use `plt.savefig()` to export your chart to a file.  
Call it **before** `plt.show()` — after `show()` the figure is cleared.

````python
x = [1, 2, 3, 4, 5]
y = [10, 24, 18, 35, 30]

plt.plot(x, y, marker="o", color="steelblue")
plt.title("My Chart")
plt.xlabel("x")
plt.ylabel("y")

# Save before show()!
plt.savefig("my_chart.png", dpi=150, bbox_inches="tight")

# bbox_inches="tight" - crops empty whitespace around the figure

plt.show()
print("Chart saved as my_chart.png")
````

## 16. Plotting with pandas + Matplotlib

pandas DataFrames have a built-in `.plot()` method that uses Matplotlib under the hood.  
This is a convenient shortcut for quick visualization — but you can still customize with `plt`.

````python
# Create a sample DataFrame
df = pd.DataFrame({
    "Month":   ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Math":    [72, 80, 68, 90, 85, 95],
    "Reading": [60, 75, 80, 70, 88, 82]
})

df.set_index("Month", inplace=True)   # use Month as the x-axis
df
````

````python
# Plot all numeric columns as lines
df.plot(kind="line", marker="o", figsize=(8, 4))

plt.title("Scores Over Months")
plt.ylabel("Score")
plt.show()
````

````python
# Bar chart from a DataFrame
df.plot(kind="bar", figsize=(8, 4))

plt.title("Monthly Scores")
plt.ylabel("Score")
plt.xticks(rotation=0)   # keep x-axis labels horizontal
plt.show()
````

````python
# Stacked bar chart
df.plot(kind="bar", stacked=True, figsize=(8, 4))

plt.title("Monthly Scores (Stacked)")
plt.ylabel("Combined Score")
plt.xticks(rotation=0)
plt.show()
````
