| Operator | Name                | Description                          | Example        | Result |
|----------|---------------------|--------------------------------------|----------------|--------|
| +        | Addition            | Adds two values                      | 5 + 3          | 8      |

Operator Precedence
| Priority | Operator       | Meaning                   |
|----------|----------------|---------------------------|
| 1(highest)| ()            | Parentheses               |
| 2         | **            | Exponentiation            |
| 3         | +x, -x        | Unary plus/minus          |
| 4         | *, /, //, %   | Multiplication / Division |
| 5         | +, -          | Addition / Subtraction    |

````python
# create a list
fruits = ["apple", "banana", "cherry"]
print(fruits) 
# access elements in the list
print(fruits[0]) # returns first element of the list
print(fruits[-1]) # returns last element of the list

# slice a list
print(fruits[1:3]) # returns first 2 elements of the list

# add an element to the end of the list
fruits.append("orange")
print(fruits)
fruits.insert(1, "grape") # insert an element at index 1
print(fruits)

# remove an element from the list
fruits.remove("banana")
print(fruits)
fruits.pop(0) # remove the first element of the list

# get length
print(len(fruits))

# check if item exists
print("apple" in fruits)
print("banana" not in fruits)

# change item
fruits[0] = "kiwi"
print(fruits)

import builtins
# built-in functions - numeric lists
print(builtins.sum([1, 2, 3])) # sum of a list
print(builtins.max([1, 2, 3])) # max of a list
print(builtins.min([1, 2, 3])) # min of a list
````

### enumerate()
`enumerate()` is a built-in Python function that lets you loop over a sequence while also keeping track of the index.

````python
# with enumerate()
fruits = ["apple", "banana", "orange"]

for i, fruit in enumerate(fruits):
    print(i, fruit)
````

What does for i, fruit in enumerate(fruits) mean?

Each iteration returns a tuple:
``` bash
(0, "apple")
```
Python unpacks it:
``` bash
i = 0
fruit = "apple"
```

## Break and Continue

break and continue are two control-flow tools in Python loops that help you control when to stop or skip parts of a loop (while and for).

1. `break` - stop the loop completely

````python
for i in range(1, 6):
    if i == 3:
        break
    print(i)
````

2. `continue`- skip this round, keep looping

````python
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
````
