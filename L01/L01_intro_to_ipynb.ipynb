# IPython Notebook (ipynb)

A ipynb is an interactive document made of **cells**. Each cell can contain either:
- **Markdown** — formatted text, headers, lists, images, and math
- **Code** — executable Python (or other languages)


## The Toolbar — What Each Button Does

At the top of the notebook you'll find a toolbar. Here's a quick reference:

| Button | Action |
|---|---|
|Code|Add a new code cell below|
|Markdown|Add a new markdown cell below|
|Run All| Run every cell|
|Select Kernel|Choose which Python environment runs your notebook|




### Essential Keyboard Shortcuts
| Shortcut | Action |
|---|---|
| `Shift + Enter` | Run cell, move to next |
| `Ctrl + Enter` | Run cell, stay on it |
| `A` (command mode) | Insert cell **above** |
| `B` (command mode) | Insert cell **below** |
| `D, D` (command mode) | Delete cell |
| `M` (command mode) | Change cell to **Markdown** |
| `Y` (command mode) | Change cell to **Code** |
| `Z` (command mode) | Undo cell deletion |
