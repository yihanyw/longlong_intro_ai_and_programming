### Type Convertion

````python
# can convert between different data types
````

- int -> float
    - python does not round anything, just adds a decimal
- float -> int
    - python does not round, it truncates (cuts off the decimal part)
    - if want rounding, use `round()`
- int -> string
    - number becomes a string
- string -> int/float
    - possible when the string is a valid number
- int -> bool
    - in python, `0` -> `False`, any non-zero -> `True`

- bool -> int
    - `True` become `1`, `False` become `0` 
- bool -> float
    - `True` become `1.0`, `False` become `0.0`

### Arithmetic Operators

| Operator | Name                | Description                          | Example        | Result |
|----------|---------------------|--------------------------------------|----------------|--------|
| +        | Addition            | Adds two values                      | 5 + 3          | 8      |
| -        | Subtraction         | Subtracts right from left            | 5 - 3          | 2      |
| *        | Multiplication      | Multiplies two values                | 5 * 3          | 15     |
| /        | Division            | Divides and returns float result     | 5 / 2          | 2.5    |
| //       | Floor Division      | Divides and rounds down              | 5 // 2         | 2      |
| %        | Modulus             | Remainder of division                | 5 % 2          | 1      |
| **       | Exponentiation      | Power operation                      | 5 ** 2

Operator Precedence
| Priority | Operator       | Meaning                   |
|----------|----------------|---------------------------|
| 1(highest)| ()            | Parentheses               |
| 2         | **            | Exponentiation            |
| 3         | +x, -x        | Unary plus/minus          |
| 4         | *, /, //, %   | Multiplication / Division |
| 5         | +, -          | Addition / Subtraction    |

## Comparison Operator

| Operator | Description              | Example   | Result |
|----------|--------------------------|------------|--------|
| `==`     | Equal to                 | `5 == 5`   | `True` |
| `!=`     | Not equal to             | `5 != 3`   | `True` |
| `>`      | Greater than             | `5 > 3`    | `True` |
| `<`      | Less than                | `5 < 3`    | `False` |
| `>=`     | Greater than or equal to | `5 >= 5`   | `True` |
| `<=`     | Less than or equal to    | `3 <= 5`   | `True` |

Comparison operators on strings compare them lexicographically (dictionary/alphabetical order), character by character.
- Python compares from left to right
- Strings use Unicode/ASCII values internally.
- Later letters in the alphabet usually have larger values
- Uppercase letters have smaller Unicode values than lowercase letters
